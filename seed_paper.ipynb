```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0, scale=3, size=1000)
hist, n_edges = np.histogram(data, bins=100)


n = (50/100) * len(hist)
perc_n = sorted(hist)[int(n)]

perc_50 = np.percentile(hist, 50)
print(perc_50, perc_n)
ind_perc = (hist <= perc_50)

plt.bar(n_edges[1:], hist, n_edges[1]-n_edges[0])
plt.bar(n_edges[1:][ind_perc], hist[ind_perc], n_edges[1]-n_edges[0])
plt.show()
```
